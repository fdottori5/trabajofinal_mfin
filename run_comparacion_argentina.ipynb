```python
import os
import sys
import subprocess

base_path = os.getcwd()
```

```python

# si le cambiaste el nombre al script, dejalo como graficos_finales.py
script_path = os.path.join(base_path, "comparacion_modelo_vs_argentina_desde_raw.py")

# output del meta-modelo
model_monthly_path = os.path.join(base_path, "Outputs", "26_profile_backtest_monthly.csv")

# csv con MERVAL, UVA, BADLAR, MPFONDO, MEP
argentina_csv_path = os.path.join(base_path, "precios_argentina.csv")

# carpeta donde querés que se guarde la comparación
comparison_output_dir = os.path.join(base_path, "comparacion_argentina")
os.makedirs(comparison_output_dir, exist_ok=True)

# chequeo rápido
print("script:", os.path.exists(script_path), script_path)
print("model monthly:", os.path.exists(model_monthly_path), model_monthly_path)
print("argentina csv:", os.path.exists(argentina_csv_path), argentina_csv_path)
print("output dir:", comparison_output_dir)

# correr el script
subprocess.run(
    [
        sys.executable,
        script_path,
        "--model-monthly",
        model_monthly_path,
        "--argentina-csv",
        argentina_csv_path,
        "--output-dir",
        comparison_output_dir,
    ],
    check=True,
)
```
